```python
import numpy as np

def generate_point_in_circle(R, n=100000):
    """
    Генерирует n точек внутри круга радиуса R.
    Возвращает массив координат (x, y).
    """
    angles = np.random.uniform(0, 2 * np.pi, n)
    radii = np.sqrt(np.random.uniform(0, 1, n)) * R
    x = radii * np.cos(angles)
    y = radii * np.sin(angles)
    return x, y

    #print() check


def compute_average_distances(R, n=100000):
    """
    Вычисляет среднее Евклидово и Манхэттенское расстояние от случайных точек до центра.
    """
    x, y = generate_point_in_circle(R, n)

    euclidean_distances = np.sqrt(x**2 + y**2)
    manhattan_distances = np.abs(x) + np.abs(y)

    avg_euclidean = round(np.mean(euclidean_distances), 2)
    avg_manhattan = np.mean(manhattan_distances)

    return avg_euclidean, avg_manhattan

# R = 1 (результаты для любого R)
R = 1
avg_euclidean, avg_manhattan = compute_average_distances(R)

avg_euclidean, avg_manhattan
```
